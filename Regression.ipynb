#Question 1:  What is Simple Linear Regression?

-> Simple Linear Regression is a statistical method used to model the relationship between two variables ‚Äî one independent variable (X) and one dependent variable (Y).

It tries to find a straight-line relationship between X and Y, which can be written as:

ùëå=ùëè0+ùëè1ùëã+ùúÄ

Y = dependent variable (the one we want to predict)

X = independent variable (the predictor)

b‚ÇÄ = intercept (value of Y when X = 0)

b‚ÇÅ = slope (change in Y for a one-unit change in X)

Œµ = error term (difference between actual and predicted Y)

 Example:

Suppose we want to predict a student's exam score (Y) based on hours studied (X).

The model might look like:

Score=40+5√ó(Hours Studied)

A student who studies 0 hours is predicted to score 40 marks (intercept).

For each extra hour studied, the score increases by 5 marks (slope).

# Question 2: What are the key assumptions of Simple Linear Regression?

-> The key assumptions of Simple Linear Regression ensure that the model gives reliable and accurate results. Here are the main ones

  1. Linearity

The relationship between the independent variable (X) and the dependent variable (Y) must be linear.

That means as X changes, Y should change in a roughly straight-line pattern.

 Example: If studying hours increase, exam scores increase at a consistent rate.

 2. Independence of Errors

The residuals (errors) should be independent of each other.

There should be no correlation between error terms.

 Example: The error for one student‚Äôs score should not depend on another student‚Äôs score.

 3. Homoscedasticity

The variance of residuals (errors) should be constant across all levels of X.

 If errors spread out more as X increases (fan shape), it‚Äôs called heteroscedasticity.

 Example: The spread of errors should look uniform across the line.

 4. Normality of Errors

The residuals should follow a normal distribution.

This is important for making valid inferences (like confidence intervals and hypothesis tests).

Example: When plotted as a histogram, residuals should look roughly bell-shaped.

 5. No or Minimal Multicollinearity

Since simple linear regression uses only one independent variable, this isn‚Äôt an issue here.

But in multiple regression, predictors should not be highly correlated with each other.

#Question 3: What is heteroscedasticity, and why is it important to address in regression models?

-> Heteroscedasticity refers to a situation in a regression model where the variance of the residuals (errors) is not constant across all levels of the independent variable(s).

In simple terms, the spread of the errors changes as the value of X changes ‚Äî the residuals ‚Äúfan out‚Äù or ‚Äúshrink‚Äù rather than staying evenly scattered.

 Example:

If you plot residuals vs. predicted values and notice that:

At lower X values, errors are small, but

At higher X values, errors become large (wider spread),
then your data shows heteroscedasticity.

# Graphically:
Homoscedasticity (Good)	Heteroscedasticity (Problem)
Errors have equal spread	Errors fan out or narrow in
Looks like random scatter	Looks like a cone or funnel shape

 Why It‚Äôs Important:

Heteroscedasticity violates a key assumption of linear regression ‚Äî that residuals have constant variance.
This can cause problems such as:

Biased standard errors ‚Üí leading to

Incorrect t-tests and F-tests

Wrong p-values

Unreliable confidence intervals and predictions

The model‚Äôs efficiency decreases, even though coefficients (like slope) may remain unbiased.

 How to Detect Heteroscedasticity:

Residuals vs. Fitted Values plot: Look for a cone/funnel pattern.

Statistical Tests:

Breusch‚ÄìPagan Test

White Test

# How to Fix It:

Transform the dependent variable (e.g., use log(Y), ‚àöY).

Use weighted least squares regression (assign smaller weights to high-variance points).

Use robust standard errors to correct for non-constant variance.

#Question 5: What is polynomial regression, and how does it differ from linear regression?

-> Polynomial Regression is a type of regression analysis that models the relationship between the independent variable (X) and the dependent variable (Y) as an n·µó ∞-degree polynomial equation.

It is used when the data shows a curved (non-linear) pattern rather than a straight-line relationship.



If we are predicting salary (Y) based on years of experience (X):

A linear regression assumes:

ùëå=ùëè0+ùëè1ùëã

‚Üí gives a straight line.

A polynomial regression might be:

ùëå=ùëè0+ùëè1ùëã+ùëè2ùëã2
	‚Äã# gives a curved line that better fits non-linear data.

> Difference Between Linear and Polynomial Regression:
Feature	Linear Regression	Polynomial Regression
Type of Relationship	Straight-line (linear)	Curved (non-linear)
Equation Form
ùëå=ùëè0+ùëè1ùëã

ùëå=ùëè0+ùëè1ùëã+ùëè2ùëã2+...+ùëèùëõùëãùëõ


* Shape of Fit	Straight line	Parabola, cubic curve, etc.

#Key Point:

It‚Äôs called ‚Äúlinear‚Äù because the equation is linear in the parameters.


#Question 6:  Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
‚óè X = [1, 2, 3, 4, 5]
‚óè Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.


```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])      # Dependent variable

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Print the regression equation
print("Intercept (b0):", model.intercept_)
print("Slope (b1):", model.coef_[0])
print(f"Regression Equation: Y = {model.intercept_:.2f} + {model.coef_[0]:.2f}X")

# Plot the data and regression line
plt.scatter(X, Y, color='blue', label='Actual Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

#Question 8:  Implement polynomial regression on the following data:
‚óè X = [1, 2, 3, 4, 5]
3
‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve.

```python
 #Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the Polynomial Regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
Y_pred = model.predict(X_poly)

# Print the model coefficients
print("Intercept (b0):", model.intercept_)
print("Coefficients (b1, b2):", model.coef_)
print(f"Equation: Y = {model.intercept_:.2f} + ({model.coef_[1]:.2f})*X + ({model.coef_[2]:.2f})*X¬≤")

# Plot the original data points
plt.scatter(X, Y, color='blue', label='Actual Data')

# Plot the polynomial regression curve
plt.plot(X, Y_pred, color='red', linewidth=2, label='Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (2nd Degree)')
plt.legend()
plt.grid(True)
plt.show()




```

#Question 9: Create a residuals plot for a regression model trained on this data:
‚óè X = [10, 20, 30, 40, 50]
‚óè Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit a Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate residuals (Actual - Predicted)
residuals = Y - Y_pred

# Print regression equation
print("Intercept (b0):", model.intercept_)
print("Slope (b1):", model.coef_[0])
print(f"Regression Equation: Y = {model.intercept_:.2f} + {model.coef_[0]:.2f}X")

# Print residuals
print("\nResiduals:")
print(residuals)

# Plot residuals
plt.scatter(X, residuals, color='blue', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Zero Error Line')
plt.xlabel('X')
plt.ylabel('Residuals (Y - Y_pred)')
plt.title('Residuals Plot to Check Heteroscedasticity')
plt.legend()
plt.grid(True)
plt.show()
```

#Question 10: Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.



->Addressing Heteroscedasticity

Heteroscedasticity means that the variance of residuals (errors) is not constant across all levels of the independent variables ‚Äî which violates regression assumptions and affects the reliability of standard errors.

Steps to fix heteroscedasticity:

Visual Inspection:
Plot residuals vs. predicted values. If the spread increases or decreases, heteroscedasticity is present.

Apply Data Transformation:

Use log, square root, or Box-Cox transformations on the dependent variable (e.g., log(Price) instead of Price).

This often stabilizes variance.

Use Weighted Least Squares (WLS):

Assign lower weights to data points with higher variance to reduce their influence.

Use Robust Standard Errors:

Apply heteroscedasticity-consistent (robust) standard errors (like White‚Äôs standard errors) to make hypothesis tests more reliable.

2. Addressing Multicollinearity

Multicollinearity occurs when two or more independent variables are highly correlated, leading to unstable coefficient estimates and inflated standard errors.

Steps to fix multicollinearity:

Detect Multicollinearity:

Calculate Variance Inflation Factor (VIF) for each variable.

If VIF > 5 (or 10), multicollinearity is high.

Examine correlation matrix among features.

Remove or Combine Variables:

Drop one of the highly correlated predictors (e.g., if area and number of rooms are strongly correlated).

Combine correlated variables into a single feature (e.g., total living area per room).

Use Regularization Techniques:

Apply Ridge Regression (L2) or Lasso Regression (L1) to penalize large coefficients and reduce collinearity effects.

Use Principal Component Analysis (PCA):

Transform correlated variables into uncorrelated components before fitting the model.

3. Ensuring a Robust Model

Split data into train and test sets to avoid overfitting.

Perform cross-validation to evaluate model performance.

Monitor metrics like R¬≤, RMSE, and Adjusted R¬≤.

Re-check residual plots and VIF after corrections to confirm issues are resolved.
